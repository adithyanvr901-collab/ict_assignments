```python
# Ultra Simple Exam Mode Game

def play_game():
    print("=== WELCOME TO THE SIGNAL GAME ===\n")

    score = 0  # start score

    # Decision 1
    print("You see a strange signal on your screen.")
    choice1 = input("Do you look at it or ignore it? (look/ignore): ").strip().lower()
    if choice1 == "look":
        print("You found an important message!")
        score += 10
    else:
        print("You ignored it and missed something important.")
        score -= 5

    # Decision 2
    print("\nThe message can be saved or deleted.")
    choice2 = input("Do you save it or delete it? (save/delete): ").strip().lower()
    if choice2 == "save":
        print("You saved the message successfully!")
        score += 5
    else:
        print("The message is lost forever.")
        score -= 5

    # Ending
    print("\n=== END ===")
    if score >= 10:
        print("Congratulations! You did a great job. ðŸŽ‰")
    else:
        print("Better luck next time! ðŸ˜„")

    print(f"Your final score: {score}\n")


# Run the game once
play_game()
```

# ***QUESTION  NUMBER  3***

```python
# ULTRA-SIMPLE CAMPUS STORE SYSTEM

# Inventory dictionary: product_name -> [price, stock]
inventory = {}

# 1. Admin adds products
n = int(input("How many products to add? "))
for _ in range(n):
    name = input("Product name: ")
    price = float(input("Price: "))
    stock = int(input("Stock: "))
    if price < 0 or stock < 0:
        print("Invalid input, skipping this product.")
    else:
        inventory[name] = [price, stock]

# 2. Customer buys products
cart = ()
while True:
    print("\nAvailable products:")
    for p, d in inventory.items():
        print(f"{p}: Price={d[0]} Stock={d[1]}")
    prod = input("Enter product to buy (or 'done'): ")
    if prod.lower() == "done":
        break
    if prod not in inventory:
        print("Product not found.")
        continue
    qty = int(input("Quantity: "))
    if qty <= 0 or qty > inventory[prod][1]:
        print(f"Invalid quantity. Available: {inventory[prod][1]}")
        continue
    cart.append((prod, qty))
    inventory[prod][1] -= qty
    if inventory[prod][1] <= 5:
        print(f"âš ï¸ Low amount of stock for {prod}!")

# 3. Generate Bill
subtotal = sum(inventory[n][0]*q for n,q in cart)
discount = 0.10*subtotal if subtotal>=5000 else 0.05*subtotal if subtotal>=2000 else 0
tax = 0.05*(subtotal-discount)
total = subtotal - discount + tax

print("\n--- BILL ---")
for n,q in cart:
    print(f"{n} x{q} = {inventory[n][0]*q:.2f}")
print(f"Subtotal: {subtotal:.2f}")
print(f"Discount: {discount:.2f}")
print(f"Tax: {tax:.2f}")
print(f"Total: {total:.2f}")
```
